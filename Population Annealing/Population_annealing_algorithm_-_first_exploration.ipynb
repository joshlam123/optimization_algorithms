```python
import numpy as np
import matplotlib.pyplot as plt
import PopulationAnnealing as PA
```

```python
n_replicas = 10
beta = np.linspace(0,1)

# Generate a trial landscape
x = np.linspace(-15,15,1000)

energies = [PA.energy(i) for i in x]
plt.plot(x,energies)
plt.show()
```

```python
# Random intital solutions
posns = np.random.choice(x,n_replicas)
posns_old = np.array(posns)
```

```python
# Resample the positions 
posns1 = PA.Resample(n_replicas,beta[0],beta[1],energies,posns)
```

```python
# Now do MCMC step (similar to SA)
n_steps = 10
posns2 = PA.MCMC_Step(x,n_steps,n_replicas,posns1,beta[1])
```

```python
# Histogram of initial posns of replicas
plt.hist(posns_old,bins=50)
plt.show()
```

```python
# Histogram of posns of replicas after resampling
plt.hist(posns1,bins=50)
plt.show()
```

```python
# Histogram of posns of replicas after MCMC
plt.hist(posns2,bins=50)
plt.show()
```
